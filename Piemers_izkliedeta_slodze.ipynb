Vēlams sākt ar balstu reakciju noteikšanu. Šim nolūkam izkliedētu slodzi $q$ aizvieto ar koncentrētu $F$.

Apskata siju divos posmos - AB un BC.

Koncentrētas slodzes (spēka) iedarbi iezīmē katra posma vidū.

$F_1=q \cdot a$

$F_2=q \cdot b $

Iezīmē balstu reakcijas $R_B$ un $R_C$, pretējā virzienā spēkiem $F_1$ un $F_2$.



Visu spēku un momentu summa ir vienāda ar 0, jo pieņem, ka sija slodzi iztur un telpā nepārvietojas.

To pieraksta:  $ \sum F_i=0 $ un $ \sum M_i=0 $

Tāpat vajadzētu būt, ka momentu summa balstos B un C ir vienāda ar nulli. Šo pieņēmumu var lietot balstu reakciju skaitliskās vērtības atrašanai.

Visu momentu summa balstā B ir vienāda ar 0.

$ \sum M_{Bi}=0 $

Moments ir spēka reizinājums ar plecu. Momentus var ieraudzīt sekojošā attēlā. Jāievēro arī virziens. Var pieņemt, ka moments pulksteņrādītāja virzienā ir negatīvs, pretējā virzienā pozitīvs.



Iezīmētos momentus summē, un izsaka nezināmo lielumu - balstu reakciju $R_C$.

$F_1 \cdot d-F_2 \cdot e+R_C \cdot (e+f)=0$

$R_C=\frac {-F_1 \cdot d+ F_2 \cdot e}{e+f}$

Tagad var veidot aprēķinu kodu *Python* sintaksē.

```python
q=120
a=3
b=6

c=a/2
d=a/2
e=b/2
f=b/2

F1=q*a
F2=q*b

Rc=(-F1*d+F2*e)/(e+f)

print (Rc)
```

Līdzīgi atrod momentus balstā C.

Visu momentu summa balstā C ir vienāda ar 0, 

$ \sum M_{Ci}=0 $



Iezīmētos momentus summē, un izsaka nezināmo lielumu - balstu reakciju $R_B$.

$F_1 \cdot (d+e+f)-F_2 \cdot f-R_B \cdot (e+f)=0$

$R_B=\frac {F_1 \cdot (d+e+f)+ F_2 \cdot f}{e+f}$

Tagad var veidot aprēķinu kodu *Python* sintaksē.


```python
Rb=(F1*(d+e+f)+F2*f)/(e+f)

print (Rb)
```

Pārbaudes aprēķins: visu spēku summa ir vienāda ar 0.

$ \sum F_i=0 $

Summējot ievēro spēku virzienu. Piemēram, uz augšu pozitīvs, uz leju negatīvs.

```python
summa=-F1-F2+Rb+Rc

print (summa)
```

Summa ir nulle, pārbaude veiksmīga.

Tagad jānoskaidro ***šķērspēka*** sadalījuma vērtības visā sijas garumā.

Sākumā var nosacīti atdalīt daļu no sijas posmā AB, sākot no kreisās puses. 

Šķēlums **s** veidots tieši pirms balsta B. Nošķelto sijas daļu BC tad var aizvietot ar šķērspēku $V_{AB'}$ un momentu $M_{AB'}$.




Vēl var ieviest mainīgu sijas garumu $x_1=x_1 \ge 0 ; x_1 \le  a$.





Spēki un momenti, kā allaž, ir līdzsvarā.

Jāpieņem pozitīvā un negatīvā zīme spēkiem un momentiem. Var pieņemt, ka spēks uz augšu ir pozitīvs, uz leju negatīvs.

Savukārt moments - pretēji pulksteņa rādītāja virzienam negatīvs, uz otru pusi attiecīgi pozitīvs.

Tad var rakstīt:

$-V_{AB'}-q \cdot x_1 = 0$

Izsakot $V_{AB'}$

$V_{AB'} = -q \cdot x_1$


```python
import numpy as np  # importē NumPy funkciju bibliotēku
from matplotlib import pyplot as plt # importē Matplotlib funkciju bibliotēku grafiku veidošanai
import matplotlib.ticker as ticker # importē Matplotlib funkciju bibliotēku grafiku asu veidošanai

S1=int(a*1000) #izveido sijas vektoru ar 1000 punktiem uz metru.
x1=np.linspace(0,a,S1)  # izveido sijas posma AB garuma vektoru. 

V_AB=-q*x1 #šķērspēka aprēkīns sijas posmam AB

plt.plot(x1,V_AB)
```

Lieces momentu aprēķina, sākumā pārveidojot izkliedētu slodzi kā koncentrētu spēku, kas pielikts aprēķināmā posma vidusdaļā.



Tad var izveidot momentu līdzsvara vienādojumu:

$M_{AB'} + q \cdot x_1 \cdot \frac {x_1} {2} = 0$

$M_{AB'} = -q \cdot x_1 \cdot \frac {x_1} {2}$

```python
M_AB=-q*x1*(x1/2) #lieces momenta aprēkīns sijas posmam AB

plt.plot(x1,M_AB)
```

Turpinājumā var nosacīti atdalīt daļu no sijas posmā BC. 

Šķēlums **s** veidots tieši pirms balsta C. Nošķelto sijas daļu tad var aizvietot ar šķērspēku $V_{BC'}$ un momentu $M_{BC'}$.



Var ieviest mainīgu sijas garumu $x_2=x_2 \ge x_1 ; x_2 \le  b$. 

$X_2$ jābūt kā $X_1$ turpinājumam - šajā piemērā jāsākas ar 3 m.

Sākumā var aprēķināt sķērsspēku $V_{BC'}$

Tad var rakstīt:

$-V_{BC'} +R_{B}-q \cdot x_2 = 0$

Izsakot $V_{BC'}$

$V_{BC'} = R_{B}-q \cdot x_2$

```python
S2=int(b*1000) #izveido sijas vektoru ar 1000 punktiem uz metru.
x2=np.linspace(a,a+b,S2) # izveido sijas posma BC garuma vektoru. Sākums no posma AB beigu punkta.

V_BC=Rb-q*x2 #šķērspēka aprēkīns sijas posmam BC

plt.plot(x2,V_BC)
```

Lieces momentu aprēķina, pārveidojot izkliedētu slodzi kā koncentrētu spēku, kas pielikts aprēķināmā posma vidusdaļā.

Lai gan aprēķini tiek veikti sijas posmam BC, aprēķinos iesaistīts ir viss sijas garums.



Tad var izveidot momentu līdzsvara vienādojumu:

$M_{BC'} + q \cdot x_2 \cdot \frac {x_2} {2} -R_B \cdot (x_2-a) = 0$

$M_{BC'} =R_B \cdot (x_2-a) - q \cdot x_2 \cdot \frac {x_2} {2} $

```python
M_BC=Rb*(x2-a)-q*x2*(x2/2)

plt.plot(x2,M_BC)
```

Otrs, iespējams, vieglāks ceļš šādā gadījumā var būt, šķeļot siju pirms balsta $R_b$ no kreisās puses.



Var ieviest mainīgu sijas garumu $x_2=x_2 \ge 0 ; x_2 \le  b$.

Vektoram $x_1$ sākumpumkts ir kreisajā pusē, pirms balsta C.

Sākumā var aprēķināt sķērsspēku $V_{CB'}$

Tad var rakstīt:

$V_{CB}+R_C-q \cdot x_2=0$

$V_{CB}= q \cdot x_2-R_C=0$



```python
x2r=np.linspace(b,0,S2) #izveido sijas posma BC garuma vektoru, kurš sākas ar maksimālo vērtību, un beidzas ar 0.

V_CB=q*x2r-Rc #aprēķina šķērspēku

plt.plot(x2,V_CB) #epīru veidojot, izvēlas horizontālās ass datus kā regulāri aprēķināto sijas posma BC garuma vektoru
```

Tad izveido momentu līdzsvara vienādojumu:

$-M_{CB'} + R_C \cdot x_2 - q \cdot x_2 \cdot \frac {x_2} {2} = 0$

$M_{CB'} = R_C \cdot x_2 - q \cdot x_2 \cdot \frac {x_2} {2} $

```python
M_CB=Rc*x2r-q*x2r*(x2r/2)

plt.plot(x2,M_CB)
```

Turpinājumā lieces momenta un šķērspēka diagrammu (epīru) veidošanas kods.

```python
V=np.append(V_AB,V_BC) #apvieno lieces momenta M, šķērspēka V un sijas garuma posmu vektorus vienā vektorā
M=np.append(M_AB,M_BC)
L=np.append(x1,x2)

#veido epīras:

fig1, ax1 = plt.subplots(1,1)

tick_spacing = 100

ax1.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

def annot_max(L,V, ax=None):
    xmax = L[np.argmax(V)]
    ymax = V.max()
    text= "V={:.0f} N, L={:.1f} m".format(ymax, xmax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.7,0.96), **kw)

annot_max(L,V)

plt.plot(L,V)
plt.plot(L, color='black')
#Fill under the curve
plt.fill_between(
        x= L, 
        y1= V, 
        #where= (0 < L)&(L < 9),
        color= "g",
        alpha= 0.2)
plt.xlim([0, np.max(L)])
plt.ylim([np.min(V)-50, np.max(V)+50])
plt.xlabel(r'Garums, m', fontsize=16)
plt.ylabel('Šķērspēks, N', fontsize=16)      
plt.show()


fig2, ax2 = plt.subplots(1,1)

tick_spacing = 100

ax2.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
def annot_max(L,M, ax=None):
    xmax = L[np.argmax(M)]
    ymax = M.max()
    text= "M={:.0f} Nm, L={:.1f} m".format(ymax, xmax)
    if not ax:
        ax2=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax2.annotate(text, xy=(xmax, ymax), xytext=(0.6,0.96), **kw)

annot_max(L,M)

ax2.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))

def annot_min(L,M, ax=None):
    xmin = L[np.argmin(M)]
    ymin = M.min()
    text= "M={:.0f} Nm, L={:.1f} m".format(ymin, xmin)
    if not ax:
        ax2=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=10")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="left", va="bottom")
    ax2.annotate(text, xy=(xmin, ymin), xytext=(0.4,0.1), **kw)

annot_min(L,M)


plt.plot(L,M)
plt.plot(L, color='black')
#Fill under the curve
plt.fill_between(
        x= L, 
        y1= M, 
        #where= (0 < L)&(L < 9),
        color= "b",
        alpha= 0.2)
plt.xlim([0, np.max(L)])
plt.ylim([np.min(M)-50, np.max(M)+50])
plt.xlabel('Garums, m', fontsize=16)
plt.ylabel('Lieces moments, Nm', fontsize=16)         
plt.show()
```
